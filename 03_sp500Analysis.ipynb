# Price / Earning OR Price / Earning per Share :

The ratio for valuing a company that measures its current share price relative to its earnings per share (EPS)


# Dividend Yield :

Companies pay dividends when they distribute some of their earnings to shareholders.

Dividends can be paid in cash or additional shares of the company's stock

# Formula : 

Dividend Yield Ratio = Dividend Per Share/Market Value Per Share


# Earning / Share OR EPS:

EPS = Net Income / Total Number of Outstanding Share

-> Outstaning share are those share that are not repurchased

# Book value :

Book value/Share = Total Equity/Number of Outstanding Share
